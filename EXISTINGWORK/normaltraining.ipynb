```python
from setfit import AbsaModel

model = AbsaModel.from_pretrained("BAAI/bge-small-en-v1.5")
```

```python
# The training/eval dataset must have `text`, `span`, `label`, and `ordinal` columns
from datasets import load_dataset

dataset = load_dataset("tomaarsen/setfit-absa-semeval-restaurants", split="train")
# dataset = load_dataset("tomaarsen/setfit-absa-semeval-restaurants", split="train")
train_dataset = dataset.select(range(409))
eval_dataset = dataset.select(range(409, 512))
```

```python
from transformers import AdamW

def custom_optimizer(model):
    return AdamW(model.parameters(), lr=3e-5, weight_decay=0.01)
from setfit import AbsaTrainer, TrainingArguments

args = TrainingArguments(
    output_dir="models",
    num_epochs=7,
    use_amp=True,
    batch_size=128,
    eval_strategy="steps",
    eval_steps=5000,
    save_steps=5000,
    load_best_model_at_end=True,
)

trainer = AbsaTrainer(
    model,
    args=args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
)
trainer.train()
```

```python
from setfit import AbsaTrainer, TrainingArguments
from transformers import EarlyStoppingCallback

args = TrainingArguments(
    output_dir="models",
    num_epochs=7,
    use_amp=True,
    batch_size=128,
    eval_strategy="steps",
    eval_steps=5000,
    save_steps=5000,
    #max_steps=50,  # Manually limit steps to 50
    load_best_model_at_end=True,
)

trainer = AbsaTrainer(
    model,
    args=args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=5)],
)
trainer.train()
```

```python
metrics = trainer.evaluate(eval_dataset)
print(metrics)
```

```python
model.save_pretrained(
    "models/setfit-absa-model-aspect",
    "models/setfit-absa-model-polarity",
)
```
